```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

## datos promedios

```python
data = pd.read_csv('data_3000.csv', sep=',', decimal='.')

data.head()
```

```python
data['vel_mean'] = data['vel_mean'].apply(lambda x: x*10)
data.head()
```

```python

#add a mean line 

plt.plot(data['vel_mean'], color='blue', linestyle='dashed', linewidth=1)
plt.xlabel('Cantidad de autos')
plt.ylabel('Velocidad promedio (m/s)')
plt.title('Velocidad promedio de los autos a lo largo del tiempo')
plt.grid(True)  # Add grid lines
plt.show()
```

```python
# Assuming you have already loaded your data into a DataFrame named 'data'
velocities = data['vel_mean']

# Create a histogram with KDE (Kernel Density Estimation) plot
sns.histplot(velocities, bins=20, kde=True, color='red', label='vel_mean')

plt.xlabel('Velocidad promedio (m/s)')
plt.ylabel('Frecuencia')
plt.title('Histograma de velocidades promedio con el mejor ajuste')
plt.legend()
plt.show()
```

```python
velocities = data['react_time']

# Create a histogram with KDE (Kernel Density Estimation) plot
sns.histplot(velocities, bins=20, kde=False, color='lightblue')

plt.xlabel('reaction time (segs)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Reaction time per driver')
plt.legend()
plt.show()
```

```python
# Assuming you have already loaded your data into a DataFrame named 'data'
velocities = data['vel_mean']
cant_autos = data['cant_autos_en_frame']

plt.figure(figsize=(10, 6))
plt.bar(cant_autos, velocities, width=3)

plt.xlabel('Cantidad de Autos al Terminar')
plt.ylabel('Velocidad Promedio (m/s)')
plt.title('Histograma de Velocidades Promedio por Cantidad de Autos')

plt.xticks(cant_autos)  # Para mostrar todas las cantidades en el eje x
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

## velocidades puntuales por auto

```python
velocidades = pd.read_csv('velocidades.csv', sep=',', decimal='.')
velocidades.head()
```

```python
velocidades = pd.read_csv('velocidades_3000.csv', sep=',', decimal='.')
velocidades = velocidades.apply(lambda x: x*10)
velocidades.head()
```

```python


def generar_color_aleatorio():
    r = random.random()
    g = random.random()
    b = random.random()
    return (r, g, b)

num_colores = 20  # Un color aleatorio por cada fila en el DataFrame
colores = [generar_color_aleatorio() for _ in range(num_colores)]

# Crear un grfico para trazar las velocidades
plt.figure(figsize=(10, 6))  # Ajustar el tamao del grfico

for i in range(num_colores):
    tiempo = np.arange(len(velocidades.iloc[i, :])) / 10.0  # Divide el tiempo por 10
    velocidad = velocidades.iloc[i, :]  # Velocidades de una fila especfica

    # Utilizar el color aleatorio y agregar etiquetas para la leyenda
    plt.plot(tiempo, velocidad, color=colores[i])

plt.xlabel('segundos')  # Aade la unidad ajustada
plt.ylabel('Velocidad (m/s)')  # Aade la unidad ajustada
plt.title('Velocidad de los autos a lo largo del tiempo')
plt.grid(True)  # Agregar lneas de cuadrcula
plt.legend()  # Mostrar la leyenda
plt.show()
```
